```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
sns.countplot(data=tips,y='day',hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
#swarmplot
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
#boxplot
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
sns.boxplot(data=tips,y='total_bill')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,color="0",marker='<')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
#split lo une
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True,palette='pastel')
plt.show()
```

```python
sns.violinplot(data=tips,y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind='swarm')
plt.show()
```

```python
#columnas con otra variable
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind='box',col='time')
plt.show()
```

```python
sns.catplot(x="day", hue="smoker",
            col="time",
            kind="count", data=tips)

plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time')
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
plt.figure(figsize=(8,8))
#markers
markers = {"Lunch":"D", "Dinner":"s"}
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.lineplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.lineplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.show()
```

```python
#col=time
sns.relplot(data= tips, x= 'total_bill', y = 'tip',  hue= 'day', style= 'time', size='size', kind= 'scatter', col = 'time');
plt.show()
```

# **Joinplot y Pairplot**

```python
tips.head(2)
```

```python
sns.jointplot(data=tips, x='total_bill',y='tip')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill',y='tip',hue='sex',kind='hist')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill',y='tip',hue='sex',kind='kde')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill',y='tip',hue='sex',kind='hist',marginal_ticks=True)
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill',y='tip',hue='sex',kind='hist',marginal_ticks=True,
              marginal_kws=dict(bins=5,fill=True,multiple='dodge'))
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill',y='tip',hue='sex',kind='hist',marginal_ticks=True,
              marginal_kws=dict(bins=25,fill=False,multiple='dodge'))
plt.show()
```

```python
sns.pairplot(data=tips)
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex')
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='coolwarm')
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='dark',corner=True)
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='coolwarm',corner=True,kind='scatter')
plt.show()
```

# **Heatmap**

```python
tips.head(2)
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr())
plt.show()
```

```python
sns.heatmap(tips.corr(),annot=True)
plt.show()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=5)
plt.show()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=5,linecolor='black')
plt.show()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=5,linecolor='black',
            vmin=0,vmax=1)
plt.show()
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=5,linecolor='black',
            vmin=-1,vmax=1)
plt.show()
```

```python
#cbar false quita la barra
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm',linewidths=5,linecolor='black',
            vmin=0.5,vmax=1,cbar=False)
plt.show()
```
