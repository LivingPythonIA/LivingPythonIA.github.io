```python
# Crear una visualización
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill',y='tip')
plt.show()
```

```python
#hue para comparar
sns.displot(data=tips, x='total_bill',y='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill',hue='sex')
plt.show()
```

```python
#“hist”, “kde”, “ecdf”
```

```python
sns.displot(data=tips, x='total_bill',hue='sex',kind='kde')
plt.show()
```

```python
#legend no esta en el grafico
sns.displot(data=tips, x='total_bill',hue='sex',kind='kde',legend=False,palette='dark',alpha=.5)
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", hue='smoker', kind='kde',legend=False, palette='dark',alpha=.25)
plt.title('Diagrama basico')
plt.xlabel('Total de la cuenta')
plt.ylabel('Density - kde')
plt.show()
```

# **Distribuciones**

```python
tips.head()
```

STAT:
    count: show the number of observations in each bin

    frequency: show the number of observations divided by the bin width

    probability: or proportion: normalize such that bar heights sum to 1

    percent: normalize such that bar heights sum to 100

    density: normalize such that the total area of the histogram equals 1

```python
sns.histplot(data=tips, x='tip')
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=15,cumulative=True)
plt.show()
```

```python
sns.histplot(data=tips, x='tip',bins=15,hue='sex')
plt.show()
```

```python
#stat estadistica de frecuencia
sns.histplot(data=tips, x='tip',bins=15,hue='sex',stat='frequency')
plt.show()
```

```python
#stat estadistica de probabilidad
sns.histplot(data=tips, x='tip',bins=15,hue='sex',stat='probability')
plt.show()
```

```python
#stat estadistica de percentil
sns.histplot(data=tips, x='tip',bins=15,hue='sex',stat='percent')
plt.show()
```

```python
#stat estadistica de densidad
sns.histplot(data=tips, x='tip',bins=15,hue='sex',stat='density')
plt.show()
```

```python
#stat estadistica cuenta es el de default
sns.histplot(data=tips, x='tip',bins=15,hue='sex',stat='count')
plt.show()
```

```python
#multiple
sns.histplot(data=tips, x='tip',bins=15,hue='sex',multiple='stack')
plt.show()
```

```python
#multiple
sns.histplot(data=tips, x='tip',bins=15,hue='sex',multiple='layer')
plt.show()
```

```python
#multiple
sns.histplot(data=tips, x='tip',bins=15,hue='sex',multiple='fill')
plt.show()
```

```python
#multiple
sns.histplot(data=tips, x='tip',bins=15,cumulative=False,hue='sex',stat='count',multiple='dodge')
plt.show()
```
