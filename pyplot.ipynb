```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
'''
Character	color
‘b’	blue
‘g’	green
‘m’	magenta
'''
plt.plot(x,y,'y')
plt.show()
```

```python
plt.plot(x,y,'bo')
plt.show()
```

```python
''' Format Strings--
‘.’	point marker
‘,’	pixel marker
‘o’	circle marker
‘v’	triangle_down marker
‘^’	triangle_up marker
‘<’	triangle_left marker
‘>’	triangle_right marker
Line Styles--
‘-’	solid line style
‘–’	dashed line style
‘-.’	dash-dot line style
‘:’	dotted line style '''
plt.plot(x,y,'rx')
plt.show()
```

```python
plt.plot(x,y,'rs-')
plt.show()
```

```python
plt.plot(x,y,'yD:')
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
