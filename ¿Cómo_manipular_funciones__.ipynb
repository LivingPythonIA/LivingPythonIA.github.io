## Dependencias

## Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```
N = 1000

def f(x):
  return x**2;

x = np.linspace(-10,10, num=N)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
N = 1000

def f(x):
  return x**2;
#agrego c
c = 5

x = np.linspace(-5,5, num=N)
#agrego c
y = f(x) + c

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
N = 1000

def f(x):
  return x**2;
#agrego c
c = 5

x = np.linspace(-5,5, num=N)
#agrego c
y = f(x) - c

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

## Alargamientos y compresiones

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=c \cdot f(x)$ alarga la gr√°fica verticalmente en un factor de $c$.
- $y= \frac{1}{c} \cdot f(x)$ comprime la gr√°fica verticalmente en un factor de $c$.
- $y=f(c \cdot x)$ comprime la gr√°fica horizontelmente en un factor de $c$.
- $y= f(\frac{1}{c} \cdot x )$ alarga la gr√°fica horizontelmente en un factor de $c$.

```
N = 1000

def f(x):
  return np.sin(x);

c = 2

x = np.linspace(-15,15, num=N)

y = f(x)


fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
N = 1000

def f(x):
  return np.sin(x);

c = 2

x = np.linspace(-15,15, num=N)
#estan los picos en 2 -2
y = c*f(x)


fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
N = 1000

def f(x):
  return np.sin(x);

c = 2

x = np.linspace(-15,15, num=N)
#-0.5
y = (1/c)*f(x)


fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
N = 1000

def f(x):
  return np.sin(x);

c = 2

x = np.linspace(-15,15, num=N)

y = f(c*x)


fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
N = 1000

def f(x):
  return np.sin(x);

c = 2

x = np.linspace(-15,15, num=N)
#se va a expendir
y = f((1/c)*x)


fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

## Reflexiones

- $y=-f(x)$ refleja la gr√°fica respecto al eje x.
- $y=f(-x)$ refleja la gr√°fica respecto al eje y.

```
N = 1000

def f(x):
  return x**3;

x = np.linspace(-10,10, num=N)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
N = 1000

def f(x):
  return x**3;

x = np.linspace(-10,10, num=N)
y = -f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

```
N = 1000

def f(x):
  return x**3;

x = np.linspace(-10,10, num=N)
#y = -f(x)
y = f(-x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```
