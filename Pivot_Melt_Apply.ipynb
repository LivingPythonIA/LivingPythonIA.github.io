### Importar librer√≠as

```python
import pandas as pd
```

### Cargar DataFrame

```python
df_books = pd.read_csv('/content/bestsellers-with-categories.csv')
df_books.head(5)
```

# **Pivot**

```python
df_books.pivot_table(index='Author',columns='Genre',values='User Rating')
```

```python
df_books.pivot_table(index='Genre',columns='Year', values='User Rating',aggfunc='sum')
```

# **Melt**

```python
df_books[['Name','Genre']].head(5)
```

```python
df_books[['Name','Genre']].head(5).melt()
```

```python
df_books.melt(id_vars='Year',value_vars='Genre')
```

### Definir funci√≥n

```python
def two_times(value):
    return value * 2
```

### apply()

```python
df_books['User Rating'].apply(two_times)
```

```python
df_books['Rating_2'] = df_books['User Rating'].apply(two_times)
```

```python
df_books
```

```python
df_books['Rating_2'] = df_books['User Rating'].apply(lambda x : x * 3)
```

```python
df_books.head()
```

```python
df_books['Rating_2'] = df_books.apply(lambda x : x['User Rating'] * 2 if x['Genre'] == 'Fiction' else x['User Rating'],axis=1)
```
