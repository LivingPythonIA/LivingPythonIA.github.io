```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
# Se usa de forma similar a plt.figure() excepto que devuelve al mismo tiempo un objeto figura y un objeto (o un arreglo) de ejes
fig, axes = plt.subplots()

# Ahora se usa el eje, como antes, para graficar en Ã©l
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2)
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)
ax1.plot(x,np.cos(x),linestyle='dashed')
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2,linewidth=8)
ax5.plot(x,np.cos(x)**2,'b', linestyle=':')
ax6.plot(x,x**2,'teal',linestyle='-.')
ax7.plot(x,y**3,'salmon')
ax8.plot(x,x**1.5,y,'^g')
fig.tight_layout()    # Para evitar superposiciÃ³n de elementos
```
