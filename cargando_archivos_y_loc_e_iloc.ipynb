### Leer JSON

```python
import json
import requests

url = "https://static.platzi.com/media/public/uploads/hpcharactersdataraw_3d934e85-dfa4-42ec-8520-fadfbecae574.json"

# Descargar el archivo JSON localmente
response = requests.get(url)
with open("archivo.json", "w") as file:
    file.write(response.text)

# Leer el archivo JSON con pandas
data = pd.read_json("archivo.json")

# Ahora puedes trabajar con el DataFrame 'data'
```

```python
data
```

```python
# Leer el archivo JSON con pandas y convertirlo en una Serie
series_data = pd.read_json("archivo.json", typ='series')
series_data
```
