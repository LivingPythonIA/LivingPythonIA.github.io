```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

## Histograma

```python
plt.hist(data,histtype='bar')
plt.show()
```

```python
plt.hist(data,bins=10,width=2)
plt.show()
```

```python
plt.hist(data,bins=50,histtype='bar')
plt.show()
```

```python
plt.hist(data,bins=10,histtype='step')
plt.show()
```

## Boxplot

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data,vert=False,patch_artist=True,notch=True)
#vert horizontal, patch artist rellena, notch achica cintura
plt.show()
```

```python
data = np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=True)
#showfliers muestra los outliers
plt.show()
```

```python
data = np.append(data,200)
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)
#showfliers false no muestra los outliers
plt.show()
```

### Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 *  np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,s=area)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='x')
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='p')
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='<')
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='>')
plt.show()
```

```python
#alpha transparencia
plt.scatter(x,y,s=area,c=colors,marker='o',alpha=0.5)
plt.show()
```
